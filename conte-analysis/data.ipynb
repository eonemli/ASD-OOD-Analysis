```python
from pathlib import Path

# Define the directory path
folder_path = Path("/ASD/ahsan_projects/Developer/ds-analysis/ebds/registered-heatmaps")

# List all files, strip the ".nii.gz" extension, and collect into a list
filenames_no_ext = [
    p.name[:-len(".nii.gz")]
    for p in folder_path.iterdir()
    if p.is_file() and p.name.endswith(".nii.gz")
]

print(filenames_no_ext)
```

```python
len(filenames_no_ext)
```

```python
def find_by_core(case_str, available_filenames):
    """
    1) Take 'case_str' like "C0197-1-1" or "T0193-1-2".
    2) Core = everything except the first character: e.g. "0197-1-1".
    3) Search in available_filenames for any name that endswith(core).
       - If exactly one match â†’ return that filename.
       - If none â†’ return None.
       - If multiple (rare) â†’ return the first match (or handle as you see fit).
    """
    if len(case_str) < 2:
        return None
    
    core = case_str[1:]  # e.g. for "C0197-1-1" â†’ "0197-1-1"
    
    # Find all names that end with the core:
    matches = [fname for fname in available_filenames if fname.endswith(core)]
    
    if len(matches) == 1:
        return matches[0]
    elif len(matches) > 1:
        # If you want to pick a priority rule (e.g. prefer exact "CONTE" prefix),
        # you could do something like:
        #   preferred = [m for m in matches if m.startswith("CONTE")]  
        #   if len(preferred)==1: return preferred[0]
        #   else: return matches[0]
        #
        # For simplicity, just return the first:
        return matches[0]
    else:
        return None


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# (4) Apply that function to every rowâ€™s "case" column, storing result in a new
#     column called "matched_name".  Rows that fail to match become NaN.
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
conte_scanners['matched_name'] = conte_scanners['case'].apply(
    lambda c: find_by_core(c, filenames_no_ext)
)


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# (5) Now see the full DataFrame with its new "matched_name" column:
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("=== Full DataFrame ===")
print(conte_scanners.to_string(index=False))


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# (6) Finally, print out any rows where no match was found (i.e. matched_name is NaN)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
nonmatching = conte_scanners[conte_scanners['matched_name'].isna()]

print("\n=== Rows with NO match ===")



print(nonmatching.to_string(index=False))
```
